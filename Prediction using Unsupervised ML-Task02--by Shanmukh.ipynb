## Prediction using Unsupervised Machine Learning

### Task-2 : From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually. 

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### Data overview

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data = data.drop('Id',axis=1)
```

```python
data.isna().sum()
```

### Visualising the data

```python
sns.set_style('darkgrid')
sns.pairplot(data,hue='Species',palette='hls');
```

```python
X = data.drop(['Species'],axis=1)
```

```python
X.head()
```

### Finding the optimum number of clusters for k-means

```python
from sklearn.cluster import KMeans
X=data.iloc[:,[0,1,2,3]].values
wcss=[]
for i in range(1,10):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    print('k:',i,'wcss: ',kmeans.inertia_)
plt.plot(range(1,10),wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Let's take 5 as number of clusters

### Applying the KMeans to dataset

```python
kmeans = KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(X)
```

#### Predicted values

```python
y_kmeans
```

### Plotting the clusters

```python
plt.figure(figsize=(10,10))
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=80,c='red',label='Iris-setosa')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=80,c='blue',label='Iris-versicolour')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=80,c='green',label='Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='Centroids')
plt.legend();
```
